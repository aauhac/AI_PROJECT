```python
import cv2
import os

path = "./dataset/train/digital_vid"
file_list = os.listdir(path)

print (f"file_list: {file_list}")
```

```python
for i in file_list:
    filepath = path + "/" + i
    video = cv2.VideoCapture(filepath) #'' 사이에 사용할 비디오 파일의 경로 및 이름을 넣어주도록 함
    length = int(video.get(cv2.CAP_PROP_FRAME_COUNT))
    width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = video.get(cv2.CAP_PROP_FPS)
    
    print("name, length, width, height, fps")
    print(f"{i}, {length}, {width}, {height}, {fps}\n")
```

```python
def split_vid(vpath, ipath, c = 0):
    vidcap = cv2.VideoCapture(vpath)
    
    count = c

    while(vidcap.isOpened()):
        ret, image = vidcap.read()

        if not ret:
            break

        #image = cv2.resize(image, (1280, 1280))

        if int(vidcap.get(1)) % 1 == 0:
            print('Saved frame number : ' + str(int(vidcap.get(1))))
            img_name = ipath.split("/")[-1]
            cv2.imwrite(ipath + f"/{img_name}_{count}.png", image)
            count += 1
    return count
    vidcap.release()
```

```python
import re
for i in file_list:
    vpath = "./dataset/train/digital_vid/" + i
    ipath = "./dataset/train/digital_img/" + i.strip(".mp4")

    if not os.path.exists(ipath):
        os.makedirs(ipath)
        print("directory create." + i)
        if os.path.isdir(vpath):
            n = 0
            for vid in os.listdir(vpath):
                n = split_vid(f"{vpath}/{vid}", ipath, n)
        else :
            split_vid(vpath, ipath)
            print()
    else:
        print('directory already existed.' + i.strip(".mp4"))
        print()
```
